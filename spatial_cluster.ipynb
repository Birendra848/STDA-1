```python
import numpy as np  # Import NumPy for numerical operations
import pandas as pd  # Import Pandas for data manipulation
import matplotlib.pyplot as plt  # Import Matplotlib for plotting
```

```python
# Load the averaged weather data with latitude and longitude
data_path = 'weather_data/averaged_weather_data_with_lat_lon.csv'
weather_data = pd.read_csv(data_path)

# Display the first few rows of the dataframe
weather_data.head()
```

```python
weather_data.tail()
```

```python
# Check the basic information of the dataframe
weather_data.info()
```

```python
# Summary statistics of the dataframe
weather_data.describe()
```

```python
# Check for missing values
weather_data.isnull().sum()
# Handle missing values if any (e.g., fill with mean or median)
weather_data.fillna(weather_data.mean(), inplace=True)
```

```python
weather_data['District'] = weather_data['District'].str.replace('>', 'A').str.replace('@', 'A').str.replace('|', 'A')
weather_data.head()
```

```python
# Plot the distribution of each feature
weather_data.hist(bins=30, figsize=(15, 10))
plt.tight_layout()
plt.show()

# Plot the geographical distribution of the data points
plt.figure(figsize=(10, 6))
plt.scatter(weather_data['lon'], weather_data['lat'], c='blue', marker='o', alpha=0.5)
plt.title('Geographical Distribution of Weather Data Points')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
import folium
from folium.plugins import MarkerCluster

# Create a map centered around the average latitude and longitude
map_center = [weather_data['lat'].mean(), weather_data['lon'].mean()]
weather_map = folium.Map(location=map_center, zoom_start=5)

# Add points to the map
marker_cluster = MarkerCluster().add_to(weather_map)
for idx, row in weather_data.iterrows():
    folium.Marker(location=[row['lat'], row['lon']], popup=row['District']).add_to(marker_cluster)

# Display the map
weather_map
```

```python
from sklearn.preprocessing import StandardScaler

# Select features for clustering
features = weather_data[['T2M', 'WS2M', 'PRECTOTCORR', 'RH2M', 'ALLSKY_SFC_SW_DWN', 'T2M_MIN', 'T2M_MAX', 'PS']]

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Create a new DataFrame with scaled features
scaled_weather_data = pd.DataFrame(scaled_features, columns=features.columns)
scaled_weather_data['lat'] = weather_data['lat']
scaled_weather_data['lon'] = weather_data['lon']
scaled_weather_data['District'] = weather_data['District']

# Display the first few rows of the prepared data
scaled_weather_data.head()
```

```python
from sklearn.cluster import KMeans

# Initialize K-Means with a chosen number of clusters
kmeans = KMeans(n_clusters=5, random_state=42)

# Fit the model to the scaled features
kmeans.fit(scaled_features)

# Add the cluster labels to the DataFrame
scaled_weather_data['kmeans_cluster'] = kmeans.labels_

# Display the first few rows with cluster labels
scaled_weather_data.head()
```

```python
from sklearn.cluster import DBSCAN

# Initialize DBSCAN with chosen parameters
dbscan = DBSCAN(eps=0.5, min_samples=5)

# Fit the model to the scaled features
dbscan.fit(scaled_features)

# Add the cluster labels to the DataFrame
scaled_weather_data['dbscan_cluster'] = dbscan.labels_

# Display the first few rows with cluster labels
scaled_weather_data.head()
```

```python
# Plot K-Means clustering results
plt.figure(figsize=(10, 6))
plt.scatter(scaled_weather_data['lon'], scaled_weather_data['lat'], c=scaled_weather_data['kmeans_cluster'], cmap='viridis', marker='o', alpha=0.5)
plt.title('K-Means Clustering Results')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

# Plot DBSCAN clustering results
plt.figure(figsize=(10, 6))
plt.scatter(scaled_weather_data['lon'], scaled_weather_data['lat'], c=scaled_weather_data['dbscan_cluster'], cmap='viridis', marker='o', alpha=0.5)
plt.title('DBSCAN Clustering Results')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
import geopandas as gpd

# Load the shapefile
shapefile_path = 'shape_file_folder/DISTRICT_BOUNDARY.shp'
districts = gpd.read_file(shapefile_path)

# Merge the shapefile with the clustering results
merged_data = districts.merge(scaled_weather_data, left_on='district_name_column', right_on='District')

# Plot the district boundaries with K-Means clustering results
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
merged_data.boundary.plot(ax=ax, linewidth=1)
merged_data.plot(column='kmeans_cluster', ax=ax, legend=True, cmap='viridis', alpha=0.5)
plt.title('K-Means Clustering Results with District Boundaries')
plt.show()

# Plot the district boundaries with DBSCAN clustering results
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
merged_data.boundary.plot(ax=ax, linewidth=1)
merged_data.plot(column='dbscan_cluster', ax=ax, legend=True, cmap='viridis', alpha=0.5)
plt.title('DBSCAN Clustering Results with District Boundaries')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Calculate the silhouette score for K-Means clustering
kmeans_silhouette_score = silhouette_score(scaled_features, scaled_weather_data['kmeans_cluster'])
print(f'K-Means Silhouette Score: {kmeans_silhouette_score}')
```

```python
# Calculate the silhouette score for DBSCAN clustering
dbscan_silhouette_score = silhouette_score(scaled_features, scaled_weather_data['dbscan_cluster'])
print(f'DBSCAN Silhouette Score: {dbscan_silhouette_score}')
```

```python
# Save the clustering results to a CSV file
scaled_weather_data.to_csv('weather_data/clustering_results.csv', index=False)
```
